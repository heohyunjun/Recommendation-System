```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 불용어 제거
tfidf = TfidfVectorizer(stop_words='english')
```

```python
dataset.head(3)
```

```python
# 영화설명 컬럼 학습
tfidf_matrix = tfidf.fit_transform(dataset['description'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

# linear_kernel : 사이킷런에서 제공하는 문서 유사도(코사인 유사도)를 계산 할 수 있음
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python
indices = pd.Series(dataset.index, index=dataset['title']).drop_duplicates()
```

```python
indices
```

```python
idx = indices['Ganglands']
print(idx)
```

```python
list(enumerate(cosine_sim[idx]))
```

```python
#Recommendation Fucntion
def get_recommendations(title):
    idx = indices[title]

    # 영화제목 인덱스 구해서 문서유사도 리스트를 구함
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 문서 유사도 리스트를 내림차순으로 소팅
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 리스트를 1 부터 10위까지 구함
    sim_scores = sim_scores[1:11]
    
    # 출력
    movie_indices = [i[0] for i in sim_scores]
    return dataset['title'].iloc[movie_indices]
```

```python
get_recommendations('Riverdale')
```
